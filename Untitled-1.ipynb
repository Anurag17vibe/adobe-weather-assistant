```python
import requests
import pandas as pd
import os
from datetime import datetime
from datetime import datetime, timedelta

LOCATIONS = {
    "Central Delhi (CP)": (28.6315, 77.2167),
    "Dwarka": (28.5921, 77.0460),
    "Rohini": (28.7041, 77.1025),
    "Noida": (28.5355, 77.3910),
    "Gurgaon": (28.4595, 77.0266),
    "Faridabad": (28.4089, 77.3178),
    "Ghaziabad": (28.6692, 77.4538),
}
import time

def pull_location_data(name, lat, lon, start="2005-01-01", end=None):
    if end is None:
        end = (datetime.today() - timedelta(days=2)).strftime("%Y-%m-%d")
    
    url = "https://archive-api.open-meteo.com/v1/archive"
    params = {
        "latitude": lat, "longitude": lon,
        "start_date": start, "end_date": end,
        "daily": [
            "temperature_2m_max", "temperature_2m_min", "temperature_2m_mean",
            "apparent_temperature_mean",
            "relative_humidity_2m_max", "relative_humidity_2m_min", "relative_humidity_2m_mean",
            "precipitation_sum", "rain_sum", "precipitation_hours",
            "windspeed_10m_max", "windgusts_10m_max",
            "shortwave_radiation_sum", "sunshine_duration"
        ],
        "timezone": "Asia/Kolkata"
    }
    
    r = requests.get(url, params=params)
    result = r.json()
    
    if "daily" not in result:
        print(f"   ⚠️ Error for {name}: {result.get('reason', result)}")
        return None
    
    d = pd.DataFrame(result["daily"])
    d["location"] = name
    d["latitude"] = lat
    d["longitude"] = lon
    return d
   
    r = requests.get(url, params=params)
    d = pd.DataFrame(r.json()["daily"])
    d["location"] = name
    d["latitude"] = lat
    d["longitude"] = lon
    return d

print("⏳ Pulling data for all NCR locations... this may take ~3-5 minutes")
all_data = []

for name, (lat, lon) in LOCATIONS.items():
    print(f"   📍 {name}...")
    result = None
    attempts = 0
    
    while result is None and attempts < 3:
        result = pull_location_data(name, lat, lon)
        if result is None:
            attempts += 1
            print(f"      ⏳ Rate limited — waiting 65 seconds before retry ({attempts}/3)...")
            time.sleep(65)
    
    if result is not None:
        all_data.append(result)
    else:
        print(f"      ❌ Skipped {name} after 3 failed attempts")
    
    time.sleep(15)  # gap between locations to respect rate limit

df = pd.concat(all_data, ignore_index=True)
df["time"] = pd.to_datetime(df["time"])

df.to_csv("delhi_ncr_weather.csv", index=False)
with open("last_update.txt", "w") as f:
    f.write(datetime.today().strftime("%Y-%m-%d"))

print(f"\n✅ Pulled {len(df)} total rows across {df['location'].nunique()} locations")
df.head()
```

```python
from datetime import datetime, timedelta
import os

def needs_refresh():
    if not os.path.exists("last_update.txt"):
        return True
    with open("last_update.txt", "r") as f:
        last_date = datetime.strptime(f.read().strip(), "%Y-%m-%d")
    days_since = (datetime.today() - last_date).days
    print(f"📅 Last data update: {last_date.date()} ({days_since} days ago)")
    return days_since >= 1  # ← changed from 7 to 1

if needs_refresh():
    print("🔄 Data is older than 1 day — refreshing...\n")
    all_data = []
    for name, (lat, lon) in LOCATIONS.items():
        print(f"   📍 Refreshing {name}...")
        result = None
        attempts = 0
        while result is None and attempts < 3:
            result = pull_location_data(name, lat, lon)
            if result is None:
                attempts += 1
                print(f"      ⏳ Rate limited — waiting 65s ({attempts}/3)...")
                time.sleep(65)
        if result is not None:
            all_data.append(result)
        time.sleep(15)

    df = pd.concat(all_data, ignore_index=True)
    df["time"] = pd.to_datetime(df["time"])
    df.to_csv("delhi_ncr_weather.csv", index=False)
    with open("last_update.txt", "w") as f:
        f.write(datetime.today().strftime("%Y-%m-%d"))
    print("✅ Data refreshed and saved!")

else:
    print("✅ Data is fresh — loading from saved file...")
    df = pd.read_csv("delhi_ncr_weather.csv")
    df["time"] = pd.to_datetime(df["time"])
    print(f"📊 Loaded {len(df)} rows across {df['location'].nunique()} locations")
```

```python

def engineer_features(group):
    group = group.sort_values("time").copy()
    group["year"] = group["time"].dt.year
    group["month"] = group["time"].dt.month
    group["day"] = group["time"].dt.day
    group["day_of_year"] = group["time"].dt.dayofyear
    group["rain_today"] = (group["rain_sum"] > 0).astype(int)

    # Tomorrow's targets
    group["temp_mean_tomorrow"] = group["temperature_2m_mean"].shift(-1)
    group["humidity_tomorrow"] = group["relative_humidity_2m_mean"].shift(-1)
    group["rain_tomorrow"] = group["rain_today"].shift(-1)

    # Yesterday's values as features
    group["temp_mean_yesterday"] = group["temperature_2m_mean"].shift(1)
    group["humidity_yesterday"] = group["relative_humidity_2m_mean"].shift(1)

    return group

df = df.groupby("location", group_keys=False).apply(engineer_features)
df = df.dropna().reset_index(drop=True)

print(f"✅ Feature engineering done! Shape: {df.shape}")
print(f"📍 Locations: {df['location'].nunique()}")
df[["time", "location", "temperature_2m_mean", "temp_mean_tomorrow", "rain_tomorrow"]].head(10)
```

```python
from sklearn.model_selection import train_test_split

# Features now include latitude & longitude so model knows the location!
feature_cols = [
    "month", "day", "day_of_year",
    "latitude", "longitude",
    "temperature_2m_max", "temperature_2m_min", "temperature_2m_mean",
    "apparent_temperature_mean",
    "relative_humidity_2m_max", "relative_humidity_2m_min", "relative_humidity_2m_mean",
    "precipitation_sum", "rain_sum", "precipitation_hours",
    "windspeed_10m_max", "windgusts_10m_max",
    "shortwave_radiation_sum", "sunshine_duration",
    "temp_mean_yesterday", "humidity_yesterday"
]

X = df[feature_cols]
y_temp = df["temp_mean_tomorrow"]
y_humidity = df["humidity_tomorrow"]
y_rain = df["rain_tomorrow"]

# Time-based split — train on older data, test on recent years
X_train, X_test, y_temp_train, y_temp_test = train_test_split(
    X, y_temp, test_size=0.2, random_state=42, shuffle=False)
_, _, y_hum_train, y_hum_test = train_test_split(
    X, y_humidity, test_size=0.2, random_state=42, shuffle=False)
_, _, y_rain_train, y_rain_test = train_test_split(
    X, y_rain, test_size=0.2, random_state=42, shuffle=False)

print(f"✅ Training set: {X_train.shape[0]} days")
print(f"✅ Testing set:  {X_test.shape[0]} days")
print(f"📊 Features: {len(feature_cols)} (including lat/long for location awareness)")
```

```python
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
import time

print("🤖 Training 3 models on 7 NCR locations × 20 years of data...\n")

# Model 1: Temperature
start = time.time()
temp_model = RandomForestRegressor(n_estimators=200, random_state=42, n_jobs=-1)
temp_model.fit(X_train, y_temp_train)
print(f"✅ Temperature model trained ({time.time()-start:.1f}s)")

# Model 2: Humidity
start = time.time()
humidity_model = RandomForestRegressor(n_estimators=200, random_state=42, n_jobs=-1)
humidity_model.fit(X_train, y_hum_train)
print(f"✅ Humidity model trained ({time.time()-start:.1f}s)")

# Model 3: Rain
start = time.time()
rain_model = RandomForestClassifier(n_estimators=200, random_state=42, n_jobs=-1)
rain_model.fit(X_train, y_rain_train)
print(f"✅ Rain model trained ({time.time()-start:.1f}s)")

print("\n🎉 All 3 models ready!")
```

```python
from sklearn.metrics import mean_absolute_error, r2_score, accuracy_score, classification_report

temp_pred = temp_model.predict(X_test)
humidity_pred = humidity_model.predict(X_test)
rain_pred = rain_model.predict(X_test)

print("=" * 50)
print("📊 MODEL PERFORMANCE REPORT")
print("=" * 50)

temp_mae = mean_absolute_error(y_temp_test, temp_pred)
temp_r2 = r2_score(y_temp_test, temp_pred)
print(f"\n🌡️  TEMPERATURE MODEL")
print(f"   Average Error : ±{temp_mae:.2f}°C")
print(f"   R² Score      : {temp_r2:.3f}")

hum_mae = mean_absolute_error(y_hum_test, humidity_pred)
hum_r2 = r2_score(y_hum_test, humidity_pred)
print(f"\n💧 HUMIDITY MODEL")
print(f"   Average Error : ±{hum_mae:.2f}%")
print(f"   R² Score      : {hum_r2:.3f}")

rain_acc = accuracy_score(y_rain_test, rain_pred)
print(f"\n🌧️  RAIN MODEL")
print(f"   Accuracy      : {rain_acc*100:.1f}%")
print(f"\n{classification_report(y_rain_test, rain_pred, target_names=['No Rain', 'Rain'])}")
print("=" * 50)
```

```python
def predict_weather(date_str, location="Central Delhi (CP)"):
    from datetime import datetime
    
    if location not in LOCATIONS:
        print(f"❌ Unknown location! Choose from:")
        for loc in LOCATIONS:
            print(f"   • {loc}")
        return
    
    lat, lon = LOCATIONS[location]
    target_date = datetime.strptime(date_str, "%Y-%m-%d")
    month = target_date.month
    day = target_date.day
    day_of_year = target_date.timetuple().tm_yday

    # Get historical average for same time of year at that location
    loc_df = df[df["location"] == location]
    similar_days = loc_df[loc_df["day_of_year"].between(day_of_year - 3, day_of_year + 3)]

    avg_row = similar_days[feature_cols].mean()
    avg_row["month"] = month
    avg_row["day"] = day
    avg_row["day_of_year"] = day_of_year
    avg_row["latitude"] = lat
    avg_row["longitude"] = lon

    X_input = pd.DataFrame([avg_row], columns=feature_cols)

    temp_result = temp_model.predict(X_input)[0]
    humidity_result = humidity_model.predict(X_input)[0]
    rain_result = rain_model.predict(X_input)[0]
    rain_proba = rain_model.predict_proba(X_input)[0][1]

    print("=" * 50)
    print(f"📅 {date_str} — {location}")
    print("=" * 50)
    print(f"🌡️  Temperature : {temp_result:.1f}°C")
    print(f"💧 Humidity    : {humidity_result:.1f}%")
    print(f"🌧️  Rain        : {'Yes ☔' if rain_result == 1 else 'No ☀️'} ({rain_proba*100:.0f}% chance)")
    print("=" * 50)

# Test all locations!
for loc in LOCATIONS:
    predict_weather("2026-07-15", loc)
    print()
```

```python
def get_owm_current(location):
    """Get live current weather from OpenWeatherMap"""
    lat, lon = LOCATIONS[location]
    url = f"https://api.openweathermap.org/data/2.5/weather"
    params = {
        "lat": lat, "lon": lon,
        "appid": OWM_API_KEY,
        "units": "metric"
    }
    r = requests.get(url, params=params)
    data = r.json()
    if r.status_code != 200:
        return None
    return {
        "temp": data["main"]["temp"],
        "temp_max": data["main"]["temp_max"],
        "temp_min": data["main"]["temp_min"],
        "humidity": data["main"]["humidity"],
        "rain": data.get("rain", {}).get("1h", 0),
        "description": data["weather"][0]["description"].title(),
        "source": "OpenWeatherMap Live 🌍"
    }

def get_owm_forecast(location, date_str):
    """Get 5-day forecast from OpenWeatherMap"""
    lat, lon = LOCATIONS[location]
    url = f"https://api.openweathermap.org/data/2.5/forecast"
    params = {
        "lat": lat, "lon": lon,
        "appid": OWM_API_KEY,
        "units": "metric"
    }
    r = requests.get(url, params=params)
    data = r.json()
    if r.status_code != 200:
        return None

    # Find forecasts matching the target date
    target = date_str
    matching = [
        f for f in data["list"]
        if f["dt_txt"].startswith(target)
    ]

    if not matching:
        return None

    # Average across all forecasts for that day
    temps = [f["main"]["temp"] for f in matching]
    humidities = [f["main"]["humidity"] for f in matching]
    rain_total = sum(f.get("rain", {}).get("3h", 0) for f in matching)
    descriptions = [f["weather"][0]["description"] for f in matching]

    return {
        "temp": sum(temps) / len(temps),
        "temp_max": max(temps),
        "temp_min": min(temps),
        "humidity": sum(humidities) / len(humidities),
        "rain": rain_total,
        "description": descriptions[0].title(),
        "source": "OpenWeatherMap Forecast 🌍"
    }

print("✅ OpenWeatherMap functions ready!")
```

```python
import dateparser
import pandas as pd
import re
import random
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings("ignore")

conversation_memory = {"last_location": "Central Delhi (CP)"}

def extract_location(text):
    text_lower = text.lower()
    shortcuts = {
        "cp": "Central Delhi (CP)",
        "central delhi": "Central Delhi (CP)",
        "delhi": "Central Delhi (CP)",
        "dwarka": "Dwarka",
        "rohini": "Rohini",
        "noida": "Noida",
        "gurgaon": "Gurgaon",
        "gurugram": "Gurgaon",
        "faridabad": "Faridabad",
        "ghaziabad": "Ghaziabad",
    }
    for keyword, full_name in shortcuts.items():
        if keyword in text_lower:
            conversation_memory["last_location"] = full_name
            return full_name
    return conversation_memory["last_location"]

def extract_date(text):
    text_lower = text.lower()
    today = datetime.now().date()

    if "today" in text_lower:
        return today.strftime("%Y-%m-%d")
    if "day before yesterday" in text_lower:
        return (today - timedelta(days=2)).strftime("%Y-%m-%d")
    if "yesterday" in text_lower:
        return (today - timedelta(days=1)).strftime("%Y-%m-%d")
    if "day after tomorrow" in text_lower:
        return (today + timedelta(days=2)).strftime("%Y-%m-%d")
    if "tomorrow" in text_lower:
        return (today + timedelta(days=1)).strftime("%Y-%m-%d")

    months = {
        "jan": 1, "january": 1, "feb": 2, "february": 2,
        "mar": 3, "march": 3, "apr": 4, "april": 4,
        "may": 5, "jun": 6, "june": 6, "jul": 7, "july": 7,
        "aug": 8, "august": 8, "sep": 9, "september": 9,
        "oct": 10, "october": 10, "nov": 11, "november": 11,
        "dec": 12, "december": 12
    }

    pattern = r'(\d{1,2})(?:st|nd|rd|th)?\s+(?:of\s+)?(' + '|'.join(months.keys()) + r')|(' + '|'.join(months.keys()) + r')\s+(\d{1,2})(?:st|nd|rd|th)?'
    match = re.search(pattern, text_lower)
    if match:
        if match.group(1) and match.group(2):
            day = int(match.group(1))
            month = months[match.group(2)]
        else:
            day = int(match.group(4))
            month = months[match.group(3)]
        candidate = today.replace(month=month, day=day)
        if candidate > today:
            candidate = candidate.replace(year=today.year - 1)
        return candidate.strftime("%Y-%m-%d")

    parsed = dateparser.parse(text, settings={
        "PREFER_DATES_FROM": "future",
        "RELATIVE_BASE": datetime.now()
    })
    if parsed:
        return parsed.date().strftime("%Y-%m-%d")

    return today.strftime("%Y-%m-%d")

def get_actual_data(date_str, location):
    loc_df = df[df["location"] == location].copy()
    loc_df["time"] = pd.to_datetime(loc_df["time"]).dt.date.astype(str)
    row = loc_df[loc_df["time"] == date_str]
    if row.empty:
        return None
    row = row.iloc[0]
    return {
        "temp": row["temperature_2m_mean"],
        "temp_max": row["temperature_2m_max"],
        "temp_min": row["temperature_2m_min"],
        "humidity": row["relative_humidity_2m_mean"],
        "rain": row["rain_sum"],
    }

def get_predicted_data(date_str, location):
    lat, lon = LOCATIONS[location]
    date_obj = datetime.strptime(date_str, "%Y-%m-%d")
    day_of_year = date_obj.timetuple().tm_yday

    loc_df = df[df["location"] == location].copy()
    loc_df["time"] = pd.to_datetime(loc_df["time"])
    loc_df["day_of_year"] = loc_df["time"].dt.dayofyear
    loc_df["month"] = loc_df["time"].dt.month
    loc_df["day"] = loc_df["time"].dt.day
    loc_df["temp_mean_yesterday"] = loc_df["temperature_2m_mean"].shift(1)
    loc_df["humidity_yesterday"] = loc_df["relative_humidity_2m_mean"].shift(1)
    loc_df["latitude"] = lat
    loc_df["longitude"] = lon

    similar_days = loc_df[loc_df["day_of_year"].between(day_of_year - 3, day_of_year + 3)]
    avg_row = similar_days[feature_cols].mean()
    avg_row["month"] = date_obj.month
    avg_row["day"] = date_obj.day
    avg_row["day_of_year"] = day_of_year
    avg_row["latitude"] = lat
    avg_row["longitude"] = lon

    X_input = pd.DataFrame([avg_row], columns=feature_cols)
    return {
        "temp": temp_model.predict(X_input)[0],
        "humidity": humidity_model.predict(X_input)[0],
        "rain": rain_model.predict(X_input)[0],
        "rain_proba": rain_model.predict_proba(X_input)[0][1],
    }

def print_prediction(data):
    rain_str = f"Yes ☔ ({data['rain_proba']*100:.0f}% chance)" if data['rain'] == 1 else f"No ☀️ ({data['rain_proba']*100:.0f}% chance)"
    print(f"   🌡️  Temperature : {data['temp']:.1f}°C")
    print(f"   💧 Humidity    : {data['humidity']:.1f}%")
    print(f"   🌧️  Rain        : {rain_str}")
    if data['rain'] == 1 and data['rain_proba'] > 0.7:
        print(f"   💡 Tip: High chance of rain — carry an umbrella! ☂️")
    elif data['temp'] > 38:
        print(f"   💡 Tip: Very hot day — stay hydrated! 💧")
    elif data['temp'] < 12:
        print(f"   💡 Tip: Cold day — wear warm clothes! 🧣")

def warm_greeting(text_lower):
    if any(w in text_lower for w in ["hi", "hello", "hey adobe", "hey"]):
        return "Hey! 😊 Great to see you! Ask me anything about Delhi NCR weather!"
    if any(w in text_lower for w in ["how are you", "how r u", "wassup", "what's up"]):
        return "I'm doing great, always sunny in here! ☀️ What weather info can I get you?"
    if "good morning" in text_lower:
        return "Good morning! 🌅 Hope it's a great day ahead! Want to check today's weather?"
    if "good night" in text_lower:
        return "Good night! 🌙 Stay weather-ready tomorrow!"
    if "good evening" in text_lower:
        return "Good evening! 🌇 Want to know what tomorrow looks like?"
    return None

def adobe_chat(user_input):
    clean = user_input.lower()
    for phrase in ["hey adobe,", "hey adobe", "adobe,", "adobe"]:
        clean = clean.replace(phrase, "").strip()

    location = extract_location(clean)
    date_str = extract_date(clean)
    date_obj = datetime.strptime(date_str, "%Y-%m-%d")
    today = datetime.now().date()
    friendly_date = date_obj.strftime("%A, %d %B %Y")
    days_from_today = (date_obj.date() - today).days

    # ── PAST DATE → Open-Meteo CSV ──
    if date_obj.date() < today:
        data = get_actual_data(date_str, location)
        if data:
            rain_str = f"Yes ☔ ({data['rain']:.1f}mm recorded)" if data['rain'] > 0 else "No ☀️"
            print(f"\n🤖 Adobe: Actual recorded data for {location} on {friendly_date}:")
            print(f"   🌡️  Avg Temp : {data['temp']:.1f}°C  (↑{data['temp_max']:.1f} / ↓{data['temp_min']:.1f})")
            print(f"   💧 Humidity : {data['humidity']:.1f}%")
            print(f"   🌧️  Rainfall : {rain_str}")
            print(f"   📊 Source   : Open-Meteo Historical Data ✅")
        else:
            print(f"\n🤖 Adobe: No recorded data for {location} on {friendly_date}.")
            print(f"   Here's my ML prediction instead:\n")
            pred = get_predicted_data(date_str, location)
            print_prediction(pred)
            print(f"   🤖 Source : ML Model")

    # ── TODAY → OpenWeatherMap LIVE ──
    elif days_from_today == 0:
        data = get_owm_current(location)
        if data:
            rain_str = f"Yes ☔ ({data['rain']:.1f}mm)" if data['rain'] > 0 else "No ☀️"
            print(f"\n🤖 Adobe: Live weather for {location} right now:")
            print(f"   🌡️  Temp      : {data['temp']:.1f}°C  (↑{data['temp_max']:.1f} / ↓{data['temp_min']:.1f})")
            print(f"   💧 Humidity  : {data['humidity']:.1f}%")
            print(f"   🌧️  Rain      : {rain_str}")
            print(f"   🌤️  Condition : {data['description']}")
            print(f"   🌍 Source    : {data['source']}")
        else:
            print(f"\n🤖 Adobe: Couldn't fetch live data — here's my ML prediction:\n")
            pred = get_predicted_data(date_str, location)
            print_prediction(pred)
            print(f"   🤖 Source : ML Model")

    # ── TOMORROW TO 5 DAYS → OWM Forecast ──
    elif 1 <= days_from_today <= 5:
        data = get_owm_forecast(location, date_str)
        if data:
            rain_str = f"Yes ☔ ({data['rain']:.1f}mm expected)" if data['rain'] > 0 else "No ☀️"
            print(f"\n🤖 Adobe: Weather forecast for {location} on {friendly_date}:")
            print(f"   🌡️  Avg Temp  : {data['temp']:.1f}°C  (↑{data['temp_max']:.1f} / ↓{data['temp_min']:.1f})")
            print(f"   💧 Humidity  : {data['humidity']:.1f}%")
            print(f"   🌧️  Rain      : {rain_str}")
            print(f"   🌤️  Condition : {data['description']}")
            print(f"   🌍 Source    : {data['source']}")
        else:
            print(f"\n🤖 Adobe: Forecast not available — here's my ML prediction:\n")
            pred = get_predicted_data(date_str, location)
            print_prediction(pred)
            print(f"   🤖 Source : ML Model")

    # ── BEYOND 5 DAYS → ML Model ──
    else:
        pred = get_predicted_data(date_str, location)
        print(f"\n🤖 Adobe: ML Prediction for {location} on {friendly_date}:")
        print_prediction(pred)
        print(f"   🤖 Source : ML Model (trained on 20 years of data)")

# Exit phrases
exit_phrases = [
    "quit", "exit", "bye", "goodbye", "thanks", "thank you",
    "thats all", "that's all", "done", "stop", "see you",
    "see ya", "cya", "ok thanks", "okay thanks", "cheers"
]

# Welcome banner
print("=" * 55)
print("   🌤️  ADOBE — Delhi NCR Weather Assistant")
print("=" * 55)
print("   💬 Past/Today → Real recorded data ✅")
print("   💬 Future     → ML Prediction 🤖")
print("   💬 No data?   → Falls back to ML 🔄")
print("\n   Hey! I'm Adobe 🌤️ Your personal Delhi NCR")
print("   weather buddy! Ask me anything — past,")
print("   present or future weather across NCR! 😊")
print("\n   Examples:")
print("   • Hey Adobe, how was Noida on 3rd July?")
print("   • What's the weather in Gurgaon today?")
print("   • Will it rain in Rohini next Monday?")
print("   • Bye / Thanks / Done → to exit")
print("=" * 55)

while True:
    try:
        user_input = input("\nYou: ").strip()
        if not user_input:
            continue

        if any(phrase in user_input.lower() for phrase in exit_phrases):
            exits = [
                "Take care! Stay dry out there! ☂️",
                "Bye bye! Don't forget your umbrella! 🌂",
                "See you soon! Stay weather-ready! 🌤️",
                "Goodbye! Hope the weather treats you well! ☀️",
                "Cheers! Come back anytime for weather updates! 🌈",
            ]
            print(f"\n🤖 Adobe: {random.choice(exits)}")
            break

        greeting = warm_greeting(user_input.lower())
        if greeting:
            print(f"\n🤖 Adobe: {greeting}")
            continue

        adobe_chat(user_input)

    except KeyboardInterrupt:
        print("\n🤖 Adobe: Oops! Shutting down — bye for now! 🌤️")
        break
```
